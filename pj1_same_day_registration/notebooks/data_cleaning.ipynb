```python
import pandas as pd

# File paths (adjust if necessary)
input_file = '/Users/yzc/Desktop/EVC project/RI-Voting-Models/raw_data/survey_data.csv'
output_file = '/Users/yzc/Desktop/EVC project/RI-Voting-Models/raw_data/survey_data_cleaned.csv'


# Load the CSV file
df = pd.read_csv(input_file)


df.head(5)
```

```python
# Keep only columns from 'Q1' onward
df = df.loc[:, 'Q1':]

# Drop the first row
df = df.iloc[1:].reset_index(drop=True)

# Fill missing values in all columns with "No Response"
df = df.fillna("No Response")

df.head(5)
```

```python
# If there is an unwanted column named 'Unnamed: 22', drop it
if 'Unnamed: 22' in df.columns:
    df = df.drop('Unnamed: 22', axis=1)
```

```python
# Save the resulting DataFrame to a new CSV
df.to_csv(output_file, index=False)
```

**Q1: Ward (One-Hot)**

```python
print(df['Q1'].value_counts())
```

```python
### 1. Q1: Ward # (One-Hot)
# Allowed ward values as strings (to match CSV entries)
allowed_wards = ['1', '3', '5', '6', '7', '8', '9', '11', '12', '14', '15']
df['Q1'] = df['Q1'].astype(str)

# If the ward is not in the allowed list, assign "No Response"
df['Q1'] = df['Q1'].apply(lambda x: x if x in allowed_wards else "No Response")

# One-hot encode Q1
df = pd.concat([df, pd.get_dummies(df['Q1'], prefix='Q1')], axis=1)
```

```python
print(df['Q1'].value_counts())
```

**Q2: Age (Ordinal)**

```python
print(df['Q2'].value_counts())
```

```python
age_mapping = {
    "18-24": 1,
    "25-29": 2,
    "30-39": 3,
    "40-49": 4,
    "50-64": 5,
    "65 or older": 6,
    "No Response": 0  # Represent missing responses as 0
}
df['Q2'] = df['Q2'].map(age_mapping).fillna(0)
```

```python
print(df['Q2'].value_counts())
```

**Q3: Gender Identity (One-Hot)**

```python
print(df['Q3'].value_counts())
```

```python
allowed_genders = ["Male", "Female", "Non-binary / Other", "Prefer not to say"]
df['Q3'] = df['Q3'].astype(str)
df['Q3'] = df['Q3'].apply(lambda x: x if x in allowed_genders else "No Response")
df = pd.concat([df, pd.get_dummies(df['Q3'], prefix='Q3')], axis=1)
```

```python
print(df['Q3'].value_counts())
```

**Q4: Marital Status (Binary)**

```python
print(df['Q4'].value_counts())
```

```python
# Mapping: Married → 1, Not married → 0, and "No Response" as a separate category (-1)
def encode_marital(status):
    s = status.strip().lower()
    if s == "married":
        return 1
    elif s == "not married":
        return 0
    else:
        return -1  # Indicates no response or unexpected entry

df['Q4'] = df['Q4'].apply(encode_marital)
```

```python
print(df['Q4'].value_counts())
```

**Q5: Ethnicity (Multi-Hot)**

```python
print(df['Q5'].value_counts())
```

```python
ethnicity_options = [
    "White",
    "Black or African American",
    "Hispanic or Latino/a",
    "Asian",
    "American Indian or Alaska Native",
    "Native Hawaiian or Pacific Islander",
    "Other"
]

# For each option, create a binary column if the option is present in the response string.
for option in ethnicity_options:
    df[f'Q5_{option}'] = df['Q5'].apply(lambda x: 1 if option in x else 0)
```
